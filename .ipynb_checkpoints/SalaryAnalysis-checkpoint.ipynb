### Splitting salary column 

```python
def salary_split(data,index):
    list_min=[]
    list_max=[]
    for i in data.iloc[:,3]:
        if(i[12:16]=="Hour"):
            mi=i.split("-")[0][1:3]
            ma=i.split("-")[1].split(" ")[0][1:3]
            mi=int(mi)*40*4*12/1000
            ma=int(ma)*40*4*12/1000
        else:
            mi=i.split("-")[0].split("$")[1].split("K")[0]
            ma=i.split("-")[1].split("$")[1].split("K")[0]
        list_min.append(int(mi))
        list_max.append(int(ma))
    return list_min,list_max
```

```python
list1,list2=salary_split(data,3)
data["salary_min"]=list1
data["salary_max"]=list2
```
